```python
import re
path = '../input_d4.txt'
def read_cards_from_file(file_path):
    cards = []
    with open(file_path, 'r') as file:
        for line in file:
            parts = line.split(':')
            card_number = int(re.search(r'Card +(\d+)',parts[0].strip()).groups()[0])
            sequences = parts[1].split('|')
            sequence_1 = [int(num) for num in sequences[0].strip().split()]
            sequence_2 = [int(num) for num in sequences[1].strip().split()]
            cards.append((card_number, sequence_1, sequence_2))
    return cards

cards_data=read_cards_from_file('../input_d4.txt')

cards_data
```

```python
def sum_2_power_count_common_numbers(cards):
    result = 0
    
    for card_number, sequence_1, sequence_2 in cards:
        print(card_number)
        print(sequence_1)
        print(sequence_2)
        common_numbers = set(sequence_1).intersection(sequence_2)
        print(common_numbers)
        common_numbers_count = len(common_numbers)
        print(common_numbers_count)
        if common_numbers_count > 0:
            result += 2 ** (len(common_numbers)-1)
            
    return result

sum_2_power_count_common_numbers(cards_data)
```
