```kotlin
val emptyLine = listOf(input[0].replace(Regex(".", RegexOption.DOT_MATCHES_ALL), "."))
val newInput: List<String> = emptyLine + input + emptyLine
val digitInsideText = Regex("[^\\d]*(\\d+)[^\\d]*")
val onlyDigits = Regex("\\d+")
val justAStar = Regex("\\*",setOf(RegexOption.DOT_MATCHES_ALL))//Regex("(\\d|\\.)*\\*+(\\d|\\.)*")
val maxIndex = input[0].length - 1
fun starPositions(line: String, lineIndex: Int, rangeAndNumber: Pair<IntRange, Int>): List<Pair<Pair<Int, Int>, Int>> {
    return justAStar.findAll(line.substring(rangeAndNumber.first))
            .flatMap {
                println(" it.group : ${it.groups}")
                it.groups 
            }
            .filterNotNull()
            .distinct()
            .map { matchGroup ->
                println("found a star $matchGroup within range ${rangeAndNumber.first.first}")
                val indexOfSubstr = rangeAndNumber.first.first
                val indexOfStar = indexOfSubstr + matchGroup.range.first
                val providedStar = Pair(Pair(lineIndex, indexOfStar), rangeAndNumber.second)
                println("provide star : $providedStar")
                providedStar
            }.toList()
}

val example = listOf(
        "...350...",
        "...*....."
)
val exampleStarPositions = starPositions(example[1], 1, rangeAndNumber = Pair(IntRange(3, 5), 350))
if (exampleStarPositions.size != 1) println("test ko : exampleStarPositions = ${exampleStarPositions}")

```
